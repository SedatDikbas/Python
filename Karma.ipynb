```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Veri okuma
excel = pd.read_csv('OSB.csv')
excel.info()


data1 = np.array(excel)
np.random.shuffle(data1)
# Sıralanmış numpy dizisini pandas veri çerçevesine dönüştürün
data = pd.DataFrame(data1, columns=excel.columns)

# Karmaşık olarak sıralanmış verileri yazdırın
print(data)


print(data.isnull().sum())

data.describe()

y=data.OSB.values
x=data.drop(["OSB"], axis=1)

x = (x - x.min()) / (x.max() - x.min())


#Algoritma
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=1, max_depth=5)

#veri setini eğitim ve test olarak ayırma
from sklearn.model_selection import train_test_split
import sklearn.metrics as mt
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=1)

model.fit(x_train, y_train)
y_pred = model.predict(x_test)

from sklearn.metrics import precision_score, accuracy_score, f1_score, recall_score
print('Doğruluk: %3f' % accuracy_score(y_test, y_pred))
print('Kesinlik: %3f' % precision_score(y_test, y_pred))
print('Duyarlılık: %3f' % recall_score(y_test, y_pred))
print('F1_score: %3f' % f1_score(y_test, y_pred))
```
