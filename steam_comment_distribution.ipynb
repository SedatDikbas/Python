```python
# prompt: ratings alanı 500ün altında toplam kaç tane positive_ratings, negative_ratings var

import pandas as pd

# CSV dosyasını oku
df = pd.read_csv('data.csv')

# ratings alanı 500'ün altında olanları filtrele
df_filtered = df[df['sum_ratings'] < 500]

# Filtrelenmiş veride positive_ratings ve negative_ratings sütunlarının toplamlarını hesapla
total_positive_ratings = df_filtered['positive_ratings'].sum()
total_negative_ratings = df_filtered['negative_ratings'].sum()


# Pozitif ve negatif oranını hesapla
if total_negative_ratings != 0:  # Sıfıra bölünmeyi önlemek için kontrol et
    ratio = total_positive_ratings / total_negative_ratings
else:
    ratio = float('inf')  # Eğer total_negative_ratings 0 ise oran sonsuz olur


# Sonuçları yazdır
print(f"Toplam Positive Ratings (ratings < 500): {total_positive_ratings}")
print(f"Toplam Negative Ratings (ratings < 500): {total_negative_ratings}")
a = total_positive_ratings / (total_positive_ratings+total_negative_ratings) * 100
print(a)
print(f"Positive/Negative Ratio: {ratio}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV dosyasını oku
df = pd.read_csv('data.csv')

# Filtreleme ve analiz için aralık belirle
bins = [0, 500, 1000, 5000, 10000, 50000, 100000, float('inf')]  # Yorum aralıkları
bin_labels = [f"{bins[i]}-{bins[i+1]}" for i in range(len(bins)-1)]

# Aralıklara göre gruplama
df['comment_range'] = pd.cut(df['sum_ratings'], bins=bins, labels=bin_labels, right=False)

# Her bir aralık için pozitif oranını hesapla
results = []
for label in bin_labels:
    group = df[df['comment_range'] == label]
    total_positive = group['positive_ratings'].sum()
    total_negative = group['negative_ratings'].sum()
    total = total_positive + total_negative
    if total > 0:
        positivity_ratio = total_positive / total * 100  # Yüzde olarak pozitif oran
    else:
        positivity_ratio = 0  # Eğer toplam 0 ise oran 0
    results.append((label, total, positivity_ratio))

# Sonuçları DataFrame'e dönüştür
result_df = pd.DataFrame(results, columns=['Comment Range', 'Total Comments', 'Positivity Ratio'])

# Grafik çizimi
plt.figure(figsize=(10, 6))
plt.bar(result_df['Total Comments'], result_df['Positivity Ratio'], width=5000, alpha=0.7, color='skyblue')

plt.xlabel('Toplam Yorum Sayısı')
plt.ylabel('Pozitiflik Oranı (%)')
plt.title('Yorum Aralığına Göre Pozitiflik Oranı')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Grafiği göster
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV dosyasını oku
df = pd.read_csv('data.csv')

# Yorum sayısına göre aralıklar belirle
bins = [0, 500, 1000, 5000, 10000, 50000, 100000, float('inf')]  # Yorum aralıkları
bin_labels = [f"{bins[i]}-{bins[i+1]}" for i in range(len(bins)-1)]

# Aralıklara göre gruplama
df['comment_range'] = pd.cut(df['sum_ratings'], bins=bins, labels=bin_labels, right=False)

# Her bir aralık için toplam yorum sayısı ve pozitif oranını hesapla
result_list = []
for label in bin_labels:
    group = df[df['comment_range'] == label]
    total_positive = group['positive_ratings'].sum()
    total_negative = group['negative_ratings'].sum()
    total_comments = total_positive + total_negative
    positivity_ratio = (total_positive / total_comments * 100) if total_comments > 0 else 0
    result_list.append((label, total_comments, positivity_ratio))

# Sonuçları bir DataFrame'e dönüştür
result_df = pd.DataFrame(result_list, columns=['Comment Range', 'Total Comments', 'Positivity Ratio'])

# Grafik çizimi
plt.figure(figsize=(12, 6))
plt.bar(result_df['Comment Range'], result_df['Positivity Ratio'], color='skyblue', alpha=0.8)

# Grafik etiketleri ve ayarları
plt.xlabel('Toplam Yorum Sayısı Aralığı', fontsize=12)
plt.ylabel('Pozitiflik Oranı (%)', fontsize=12)
plt.title('Yorum Aralığına Göre Pozitiflik Oranı', fontsize=14)

# Izgara çizgileri ve eksen görünürlüğü
plt.grid(axis='y', linestyle='--', color='gray', alpha=0.7)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.gca().spines['top'].set_visible(False)  # Üst kenar çizgisini gizle
plt.gca().spines['right'].set_visible(False)  # Sağ kenar çizgisini gizle
plt.gca().spines['left'].set_visible(True)  # Sol kenar çizgisini göster
plt.gca().spines['bottom'].set_visible(True)  # Alt kenar çizgisini göster

plt.tight_layout()

# Grafiği göster
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV dosyasını oku
df = pd.read_csv('data.csv')

# Veri tipini kontrol ve dönüştür
df['sum_ratings'] = pd.to_numeric(df['sum_ratings'], errors='coerce')
df['positive_ratings'] = pd.to_numeric(df['positive_ratings'], errors='coerce')
df['negative_ratings'] = pd.to_numeric(df['negative_ratings'], errors='coerce')

# Eksik değerleri çıkar
df = df.dropna(subset=['sum_ratings', 'positive_ratings', 'negative_ratings'])

# Aralıkları belirle
bin_width = 250
bins = list(range(1000, 5000 + bin_width, bin_width))

# Gruplandır ve say
sum_counts = pd.cut(df['sum_ratings'], bins=bins).value_counts().sort_index()
positive_counts = pd.cut(df['positive_ratings'], bins=bins).value_counts().sort_index()
negative_counts = pd.cut(df['negative_ratings'], bins=bins).value_counts().sort_index()

# Sadece dolu grupları göster
sum_counts = sum_counts[sum_counts > 0]
positive_counts = positive_counts[positive_counts > 0]
negative_counts = negative_counts[negative_counts > 0]

# Bar grafiği için X pozisyonlarını ayarla
x_labels = [f"{int(interval.left)}-{int(interval.right)}" for interval in sum_counts.index]
x = range(len(x_labels))  # X ekseni için pozisyonlar

# Barların genişliği
bar_width = 0.3

# Histogram grafiğini oluştur
plt.bar(x, sum_counts.values, width=bar_width, label='Sum Ratings', color='skyblue', align='center')
plt.bar([i + bar_width for i in x], positive_counts.values, width=bar_width, label='Positive Ratings', color='green', align='center')
plt.bar([i + 2 * bar_width for i in x], negative_counts.values, width=bar_width, label='Negative Ratings', color='red', align='center')

# Grafik ayarları
plt.title("Histogram of Ratings")
plt.xlabel("Rating Ranges")
plt.ylabel("Count")
plt.xticks([i + bar_width for i in x], x_labels, rotation=45, ha='right')  # X eksenindeki etiketleri eğik yap
plt.legend()  # Grafikte bir açıklama ekle
plt.tight_layout()  # Grafik elemanlarının sıkışmasını önle
plt.show()
```

```python
# prompt: Ad.csv dosyamda name alanında geçen kelimelerin countını hesaplayacak ve yazdıracak kod en fazla olan ilk 100 taneyi ekrana yazdır bağlaçları sayıları ve özel karakterleri görmezden gel

import pandas as pd
import re

def count_words_in_name(csv_file_path):
    try:
        df = pd.read_csv(csv_file_path)
    except FileNotFoundError:
        print(f"Error: File '{csv_file_path}' not found.")
        return None

    if 'name' not in df.columns:
        print("Error: 'name' column not found in the CSV file.")
        return None

    word_counts = {}
    for name in df['name']:
        if pd.isna(name):  # Handle missing values
            continue

        # Clean the name string (remove non-alphanumeric characters and convert to lowercase)
        cleaned_name = re.sub(r'[^a-zA-Z\s]', '', str(name)).lower()

        # Split the cleaned name into words
        words = cleaned_name.split()

        # Count the occurrences of each word
        for word in words:
          if not word: #Ignore empty words
              continue
          if word in word_counts:
            word_counts[word] += 1
          else:
            word_counts[word] = 1


    # Sort words by count (descending)
    sorted_word_counts = dict(sorted(word_counts.items(), key=lambda item: item[1], reverse=True))

    # Print the top 100 words
    top_100_words = dict(list(sorted_word_counts.items())[:25])
    for word, count in top_100_words.items():
        print(f"{word}: {count}")

    return sorted_word_counts

# Example usage
word_counts = count_words_in_name('Ad.csv')
```
